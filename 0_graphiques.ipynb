```python
import matplotlib.pyplot as plt
import numpy as np

avignon = pd.read_csv('/content/drive/MyDrive/Projet_Long/Avignon /1_haies.csv')
bretagne = pd.read_csv('/content/drive/MyDrive/Projet_Long/Partie Code/1_haies.csv')


# Créer un histogramme circulaire avec des rectangles
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16,4),subplot_kw=dict(projection='polar'))

n, bins, patches = axes[0].hist(bretagne['Ang'].values, bins=48, range=(0, 2*np.pi), density=True)
# Définir les couleurs des rectangles
for i, patch in enumerate(patches):
    patch.set_facecolor(plt.cm.viridis(i/len(patches)))

n, bins, patches = axes[1].hist(avignon['Ang'].values, bins=48, range=(0, 2*np.pi), density=True)
for i, patch in enumerate(patches):
    patch.set_facecolor(plt.cm.viridis(i/len(patches)))

# Afficher le graphique
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16,4))

sns.histplot(data= bretagne, x = 'Lng', ax = axes[0])
sns.histplot(data = avignon, x = 'Lng', ax = axes[1])
```
